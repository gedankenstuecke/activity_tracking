```python
import json
import pandas as pd
```

```python
# reads a single json file so far. Would need to be adapted to read multiple files later on
d = open('healthkit-export_2017-11-01_2017-11-21_1511297617.26078.json').readlines()
```

```python
# convert lines to JSON
json_data = json.loads("".join(d))
json_data
```

```python
# extract heart rate, calories burned & step count into lists
dates = []
variables = []
values = []

for i in json_data['HKQuantityTypeIdentifierHeartRate']:
    dates.append(i['sdate'])
    variables.append('HeartRate')
    values.append(i['value'])
    
    
for i in json_data['HKQuantityTypeIdentifierStepCount']:
    dates.append(i['sdate'])
    variables.append('StepCount')
    values.append(i['value'])
    
for i in json_data['HKQuantityTypeIdentifierActiveEnergyBurned']:
    dates.append(i['sdate'])
    variables.append('ActiveEnergyBurned')
    values.append(i['value'])
```

```python
# right now the dates on the healthkit exports are broken in two places
# 1. seconds are milliseconds 
# 2. hours are in 12h format instead of 24h, but the AM/PM marker is missing. 
# this at least fixes the milliseconds somewhat nonsensical
fixed_dates = []
for i in dates:
    print (i)
    date = i.replace('+00:00','')
    fixed_dates.append(date[:-2] + str(int(int(date[-2:])/(100/60))) + "+00:00")
```

```python
# create a dataframe from the 3 columns and export to CSV
df = pd.DataFrame(data = {'date': fixed_dates,
                     'variable': variables,
                     'value' : values})
df.to_csv('healthkitdata.csv')
```
