```python
import json
import pandas as pd
import glob
from collections import defaultdict

def read_metadata(json_obj):
    metadata = json_obj['userData']
    if 'dateOfBirth' in metadata.keys():
        dob = metadata['dateOfBirth'][:10]
    else:
        dob = "-"

    if 'biologicalSex' in metadata.keys():
        sex = metadata['biologicalSex']
    else:
        sex = "-"
    return (dob,sex)

users = defaultdict(int)
data_types = defaultdict(int)

files = glob.glob("*.json")
files.sort()

metadata_found = []

metadata_file = open("results/metadata.csv","w")
metadata_file.write("user\tdateOfBirth\tbiologicalSex\n")

heartrate_file = open("results/heartrate.csv","w")
metadata_file.write("user\tunit\tsdate\tvalue\n")

stepcount_file = open("results/stepcount.csv","w")
stepcount_file.write("user\tunit\tsdate\tedate\tvalue\n")

distance_file = open("results/distance.csv","w")
distance_file.write("user\tunit\tsdate\tedate\tvalue\n")

for f in files:
    user = f.split('-')[0]
    try:
        d = open(f).readlines()
        json_data = json.loads("".join(d))
        
        # get meta data
        dob,sex = read_metadata(json_data)
        if dob != "-" and sex != "-" and user not in metadata_found:
            print('{}\t{}\t{}'.format(user,dob,sex))
            metadata_file.write('{}\t{}\t{}\n'.format(user,dob,sex))
            metadata_found.append(user)
        # get HR
        hr = json_data['HKQuantityTypeIdentifierHeartRate']
        for entry in hr:
            heartrate_file.write('{}\t{}\t{}\t{}\n'.format(
                user,
                entry['unit'],
                entry['sdate'],
                entry['value']))
            
        # get HKQuantityTypeIdentifierStepCount: {value, count, edate, sdate}
        steps = json_data['HKQuantityTypeIdentifierStepCount']
        for entry in steps:
            stepcount_file.write("{}\t{}\t{}\t{}\t{}\n".format(
                user,
                entry['unit'],
                entry['sdate'],
                entry['edate'],
                entry['value']))
        
        # get HKQuantityTypeIdentifierDistanceWalkingRunning
        distance = json_data['HKQuantityTypeIdentifierDistanceWalkingRunning']
        for entry in distance:
            distance_file.write("{}\t{}\t{}\t{}\t{}\n".format(
                user,
                entry['unit'],
                entry['sdate'],
                entry['edate'],
                entry['value']))
    except: 
        next
    
metadata_file.close()    
heartrate_file.close()   
stepcount_file.close()
distance_file.close()
```
